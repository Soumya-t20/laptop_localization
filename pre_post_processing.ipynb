### Scan all the WiFis to select usable ones.

```python
from pywifi import PyWiFi
import pandas as pd
import json
import time
from pywifi import PyWiFi
```

```python
wifis = {}
```

```python
# Scan all the WiFis.

def scan_wifis(name):
    wifi = PyWiFi()
    iface = wifi.interfaces()[0]

    print("Scanning for Wi-Fi networks...")

    access_points = iface.scan_results()
    location = {}
    for ap in access_points:
        print(f"SSID: {ap.ssid}, BSSID: {ap.bssid}, RSSI: {ap.signal} dBm")
        location[ap.bssid] = [ap.ssid, ap.signal, ap.bssid]
    wifis[f"{name}"] = list(location.values())


scan_wifis("SE")

print("\n", wifis)

with open("home_wifis_at_extremes.json", "w") as outfile: 
    json.dump(wifis, outfile)
```

```python
# Code to get Unique SSIDs and keep selected SSIDs and BSSIDs

w = {}

with open("nguyen_wifis_at_extremes.json", "r") as file:
    w = json.load(file)
    
all = []
for v in w.values():
    all+=(v)

allk = {}

for i in all:
    allk[i[2]] = i[0]

for i in allk.keys():
    print(i, allk[i])

uniq_bssids = list(allk.keys())
uniq_ssids = list(set(allk.values()))

print("\n\nUnique BSSIDs:", len(uniq_bssids), uniq_bssids)
print("\n\nUnique SSIDs:", len(uniq_ssids), uniq_ssids)
```

```python
selected_ssids = ["MASON-SECURE", "eduroam", "MASON"]
selected_bssids = []
selected_mapping = {}

# Check if selected ssids are present in every corner of the layout:
for i in w.keys():
    ssids = set()
    for j in w[i]:
        ssids.add(j[0])
    ssids = list(ssids)
    c = []
    for sss in selected_ssids:
        if sss in ssids:
            c.append(sss)
    print(i, "All Found" if len(c) == 3 else f"Only {c} are found")

# Appending BSSIDs based on SSIDs
for i in uniq_bssids:
    if allk[i] in selected_ssids:
        selected_bssids.append(i)
        selected_mapping[i] = allk[i]
        
print("\n\nSelected BSSIDs:", selected_bssids)

selected = {
    "bssids": selected_bssids,
    "ssids": selected_ssids,
    "mapping": selected_mapping
}

with open("selected.json", "w") as outfile: 
    json.dump(selected, outfile)
```
